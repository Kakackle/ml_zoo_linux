## Create dataset from files

```python
train_ds, val_ds = keras.utils.image_dataset_from_directory(
    directory = f"{ds_folder}",
    labels = "inferred",
    color_mode = "rgb",
    subset="both",
    validation_split=0.2,
    shuffle=True,
    seed=1337,
    image_size=image_size,
    batch_size=batch_size
)
```

## Data augmentation
Before engaging in data augmentation, let's count the numbers of class instances in the dataset - are there any imbalances? Is the dataset large enough?

```python
for folder in os.listdir(f"./{ds_folder}"):
    # file_count = len([name for name in os.listdir(f'./garbage_classification/{folder}') if os.path.isfile(name)])
    file_count = len(os.listdir(f'./garbage_classification/{folder}'))
    print(f'{folder} class count: {file_count}')
```

It would seem the 'clothes' class seems to heavily (over 5x) overrepresent the dataset, which, on the other hand might make sense, as the class includes multiple articles of clothes, which can vary significantly (think shoes vs shirts vs pants).

Other than that, the data could always use some augmentation, as in simulating taking the photos of the images from different angles, in different circumstances.

```python
data_augmentation_layers = [
    layers.RandomFlip("horizontal"),
    layers.RandomFlip("vertical"),
    layers.RandomBrightness(factor=0.1),
    layers.RandomRotation(factor=0.1),
    layers.RandomZoom(height_factor=0.1)
]

def data_augmentation(images):
    for layer in data_augmentation_layers:
        images = layer(images)
    return images
```

```python
# plt.figure(figsize=(10,10))
# for images, labels in train_ds.take(1):
#     indexes = []
#     for _ in range(9):
#         indexes.append(random.randint(0, batch_size))
#     augmented_images = data_augmentation(images)
#     for i, ind in enumerate(indexes):
#         ax = plt.subplot(3,3,i+1)
#         plt.imshow(np.array(augmented_images[0]).astype("uint8"))
#         plt.title(class_names[int(labels[0])])
#         plt.axis("off")
```

```python
plt.figure(figsize=(10, 10))
for images, _ in train_ds.take(1):
    for i in range(9):
        augmented_images = data_augmentation(images)
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(np.array(augmented_images[0]).astype("uint8"))
        plt.axis("off")
```

### applying augmentation to train data
Note: this doesn't actually apply these transformations now - it creates a mapping with batching and prefetching, that will be called asynchronously during retrieving the batches during the fitting process

```python
# Apply `data_augmentation` to the training images.
train_ds_aug = train_ds.map(
    lambda img, label: (data_augmentation(img), label),
    num_parallel_calls=tf.data.AUTOTUNE,
)
# Prefetching samples in GPU memory helps maximize GPU utilization.
train_ds_aug = train_ds_aug.prefetch(tf.data.AUTOTUNE)
val_ds = val_ds.prefetch(tf.data.AUTOTUNE)
```

```python
tf.data.experimental.cardinality(train_ds_aug).numpy()
```

Theoretically, the augmentations should be applied on the beginning of every epoch, making the total number of images trained to equal to 72 * 32 * number of epochs. For example for 20 epochs, it could be 46080 images, which might inadvertently cause the model to overfit on the training data, if the augmentations are not significant enough, whilst augmentations too large, may be destructive and obstructing the learning of class object features

## Modelling

### Simple conv model

```python
num_classes = len(class_names)
```

```python
model_small = tf.keras.Sequential([
    layers.Rescaling(1./255),
    layers.Conv2D(filters=64,  kernel_size = 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(filters=64,  kernel_size = 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(filters=64,  kernel_size = 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(num_classes),
    layers.Softmax()
])
```

### Important - when dealing with image (or other heavy data format) - modelcheckpoints can prove very beneficiary
Along with other callbacks like earlystopping

```python
callbacks = [
    keras.callbacks.ModelCheckpoint("model_small_three_at_{epoch}.keras"),
]
```

```python
model_small.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
              metrics=['accuracy'])
```

```python
epochs = 20
```

### GPU memory fix?

```python
# config = tf.compat.v1.ConfigProto()
# config.gpu_options.allow_growth = True
# sess = tf.compat.v1.Session(config=config)
```

```python
history = model_small.fit(
    train_ds_aug,
    validation_data=val_ds,
    epochs=epochs,
    # callbacks = callbacks
)
```

```python
history_df = pd.DataFrame(history.history)
history_df[['loss', 'val_loss']].plot()
```

```python
history_df[['accuracy', 'val_accuracy']].plot()
```

The end test accuracy of the model is well over 90%, whilst the validation accuracy seems to oscillate around 85%. Looking at the learning curves, it seems the model might be overfitting quite a bit, even though it's capacity is not that large (for a convolutional model)

```python
sample_preds = []
sample_argmax = []
for img in sample_images:
    predictions = model_small.predict(img)
    # preds = np.argmax(predictions, 1)[0]
    sample_preds.append(predictions)
    sample_argmax.append(np.argmax(predictions,1)[0])
print(sample_preds)
print(sample_argmax)
```

```python
def plot_image(i, predictions_array, true_label, img):
  true_label, img = true_label[i], img[i]
  img = tf.squeeze(img, 0)
  img = img/255
  plt.grid(False)
  plt.xticks([])
  plt.yticks([])

  plt.imshow(img, cmap=plt.cm.binary, vmin=0, vmax=255)

  predicted_label = np.argmax(predictions_array)
  if predicted_label == true_label:
    color = 'blue'
  else:
    color = 'red'

  plt.xlabel("{} {:2.0f}% ({})".format(class_names[predicted_label],
                                100*np.max(predictions_array),
                                class_names[true_label]),
                                color=color)

def plot_value_array(i, predictions_array, true_label):
  true_label = true_label[i]
  plt.grid(False)
  plt.xticks(range(10))
  plt.yticks([])
  thisplot = plt.bar(range(3), predictions_array, color="#777777")
  plt.ylim([0, 1])
  predicted_label = np.argmax(predictions_array)

  thisplot[predicted_label].set_color('red')
  thisplot[true_label].set_color('blue')
```

### Normalization with batchnorm

It seems the model confuses paper for plastic quite often, resulting in a low recall score for images of class 'paper', noticeably lowering the overall accuracy score.

Looking at the training curves, perhaps we can improve the model's learning by reducing the overfitting. Let's do that.

Also, we can add a callback that will stop the training early and restore the best performing weights, whilst increasing the maximum number of allowed epochs

For reducing overfit, we cannot introduce dropout layers as with fully-connected networks, but a batchnorm can still be utilized.

```python
model_small_batchnorm = tf.keras.Sequential([
    layers.Rescaling(1./255),
    layers.Conv2D(filters=64,  kernel_size = 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.BatchNormalization(),
    layers.Conv2D(filters=64,  kernel_size = 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.BatchNormalization(),
    layers.Conv2D(filters=64,  kernel_size = 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.BatchNormalization(),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(num_classes),
    layers.Softmax()
])
```

```python
early_stop = tf.keras.callbacks.EarlyStopping(
    monitor = "val_accuracy",
    min_delta = 0.01,
    patience = 5,
    verbose = 0,
    restore_best_weights=True
)
```

```python
callbacks = [
    # keras.callbacks.ModelCheckpoint("three_batchnorm_at_{epoch}.keras"),
    early_stop
]
```

```python
model_small_batchnorm.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
              metrics=['accuracy'])
```

```python
epochs = 25
```

```python
history = model_small_batchnorm.fit(
    train_ds_aug,
    validation_data=val_ds,
    epochs=epochs,
    callbacks = callbacks
)
```

```python
history_df = pd.DataFrame(history.history)
history_df[['loss', 'val_loss']].plot()
```

```python
history_df[['accuracy', 'val_accuracy']].plot()
```

```python
model_small_batchnorm.summary()
```

```python
test_preds = model_small_batchnorm.predict(val_ds)
test_labels = np.concatenate([y for x,y in val_ds], axis=0)
test_preds_argmax = test_preds.argmax(axis=1)

nn_clf_report = classification_report(test_labels, test_preds_argmax, target_names = class_names)
print(nn_clf_report)

nn_cm = confusion_matrix(test_labels, test_preds_argmax)
nn_cm_disp = ConfusionMatrixDisplay(nn_cm, display_labels = class_names)
nn_cm_disp.plot(xticks_rotation=-90)
```

We seem to have **overregulated** the model?
* The accuracy on training data has improved
* But the validation accuracy and loss seem even more unstable - as if the model had trouble learning more general characteristics of images / weights.

#### What can we do about this?
* Perhaps the training (augmented) data is too different from the unaugmented validation data - maybe we can try to train on unaugmented data, which will of course yield less images to train on, but in this case may prove beneficiary

### Trying an even smaller model with an additional dropout in the 'dense' classification section?

```python
model_smaller = tf.keras.Sequential([
    layers.Rescaling(1./255),
    layers.Conv2D(filters=64,  kernel_size = 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.BatchNormalization(),
    layers.Conv2D(filters=64,  kernel_size = 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.BatchNormalization(),
    layers.Flatten(),
    layers.Dropout(0.3),
    layers.Dense(64, activation='relu'),
    layers.Dense(num_classes),
    layers.Softmax()
])

model_smaller.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
              metrics=['accuracy'])

history = model_smaller.fit(
    train_ds,
    validation_data=val_ds,
    epochs=epochs,
    callbacks = callbacks
)
```
